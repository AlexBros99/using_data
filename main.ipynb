```python

from torch.utils.data import Dataset,DataLoader
import torch
```

```python
class customDataset(Dataset):
    def __init__(self,x,y):
        """Dataset with x and y values

        Args:
            x (_type_): The input features
            y (_type_): The output labels
        """
        self.x = x
        self.y = y
    
    def __len__(self):
        #Return total number of samples in dataset
        return len(self.x)
    
    def __getitem__(self, index):
        """fetch the sample at index

        Args:
            index (_type_): index from dataset

        Returns:
            _type_: the index of the sample to retrieve
        """
        return self.x[index], self.y[index]
    
#Generate Synthetic data
torch.manual_seed(0) #For reproductibility
x = torch.arange(0,100, dtype=torch.float32)
y = 2*x-1
    
#Create instance
dataset = customDataset(x,y)
    
#Use DataLoader to handle batching and shuffling
data_loader = DataLoader(dataset,batch_size=10, shuffle=True)
    
#Iterate through the DataLoader
for batch_index, (inputs,labels) in enumerate(data_loader):
    print(f"Batch {batch_index+1}")
    print("Inputs: ", inputs)
    print("Labels: ", labels)
    #Break for demonstration
    if batch_index == 0:
        break  
```
